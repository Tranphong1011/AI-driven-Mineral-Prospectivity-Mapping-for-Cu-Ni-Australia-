```python
import os

def print_tree(root_dir):
    for root, dirs, files in os.walk(root_dir):
        level = root.replace(root_dir, "").count(os.sep)
        indent = " " * 4 * level
        print(f"{indent}üìÅ {os.path.basename(root)}/")

        subindent = " " * 4 * (level + 1)
        for f in files:
            print(f"{subindent}üìÑ {f}")


```

```python
from pathlib import Path
import pandas as pd
import geopandas as gpd

deposit_dir = Path(r"C:\Users\Phong\Desktop\GIS\Project 2\4_GA_OZMIN_MineralDeposits_v02_20130830\Deposit Mineral")

csv_files = list(deposit_dir.glob("DepositData_*.csv"))
len(csv_files)
```

```python
dfs = []
for f in csv_files:
    df = pd.read_csv(f)
    df["source_file"] = f.name  # trace
    dfs.append(df)

deposits = pd.concat(dfs, ignore_index=True)
deposits.shape
```

```python
gdf_dep = gpd.GeoDataFrame(
    deposits,
    geometry=gpd.points_from_xy(
        deposits["LONG_GDA94"],
        deposits["LAT_GDA94"]
    ),
    crs="EPSG:4283"
)

gdf_dep.head()
```
