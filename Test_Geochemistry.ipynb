```python

# chn ct geochem numeric
value_cols = [
    c for c in gdf_3577.columns
    if (c.endswith("_PPM") or c.endswith("_PPB") or c.endswith("_PCT"))
    and pd.api.types.is_numeric_dtype(gdf_3577[c])
]

# -----------------------
# 1) Unit -> PPM (in-place)
# -----------------------
gdf_ppm = gdf_3577.copy()
for c in value_cols:
    if c.endswith("_PPB"):
        gdf_ppm[c] = gdf_ppm[c] / 1000.0
    elif c.endswith("_PCT"):
        gdf_ppm[c] = gdf_ppm[c] * 10000.0

# (optional) i tn AU_PPB -> AU_PPM cho r rng (khng bt buc)
# gdf_ppm = gdf_ppm.rename(columns={"AU_PPB": "AU_PPM"})

# cp nht li list nu bn i tn
# value_cols = [("AU_PPM" if c=="AU_PPB" else c) for c in value_cols]

# -----------------------
# 2) Detection limit: <=0 -> 0.5 * min_positive (per column)
# -----------------------
for c in value_cols:
    s = gdf_ppm[c].astype(float)
    min_pos = s[s > 0].min()
    if pd.notna(min_pos):
        gdf_ppm[c] = s.where(s > 0, 0.5 * min_pos)
    else:
        gdf_ppm[c] = np.nan  # ct qu rng

# -----------------------
# 3) Gi cc element  coverage
#    (khuyn ngh: >=10%  gim noise v gim survey-bias)
# -----------------------
MIN_PCT = 10.0
coverage = gdf_ppm[value_cols].notna().mean() * 100
keep_cols = coverage[coverage >= MIN_PCT].index.tolist()

# m bo lun gi core suite cho Cu (nu c)
core = ["CU_PPM", "PB_PPM", "ZN_PPM", "AS_PPM", "AU_PPB", "MO_PPM", "CO_PPM", "NI_PPM", "MN_PPM", "FE_PCT", "U_PPM", "SB_PPM", "CR_PPM"]
for c in core:
    if c in gdf_ppm.columns and c not in keep_cols:
        keep_cols.append(c)

print("Keep elements:", len(keep_cols))

# -----------------------
# 4) Deduplicate theo SAMPLEID (median cho numeric, mode cho company/method)
# -----------------------
# helper mode
def mode_or_nan(s):
    s = s.dropna()
    if len(s) == 0:
        return np.nan
    return s.value_counts().index[0]

group_cols = ["SAMPLEID"]

agg_dict = {c: "median" for c in keep_cols}
# metadata: chn mode
meta_cols = ["COMPANY", "SAM_METH", "SAM_TYPE", "ACCURACY", "MAP100K", "MAP250K"]
for mc in meta_cols:
    if mc in gdf_ppm.columns:
        agg_dict[mc] = mode_or_nan

# geometry: ly median x/y  dng li point (n cho duplicate gn nhau)
gdf_ppm["x"] = gdf_ppm.geometry.x
gdf_ppm["y"] = gdf_ppm.geometry.y
agg_dict["x"] = "median"
agg_dict["y"] = "median"

df_dedup = gdf_ppm.groupby(group_cols, as_index=False).agg(agg_dict)

gdf_points = gpd.GeoDataFrame(
    df_dedup.drop(columns=["x","y"]),
    geometry=gpd.points_from_xy(df_dedup["x"], df_dedup["y"]),
    crs=gdf_ppm.crs
)

print("Points after dedup:", len(gdf_points))

# -----------------------
# 5) Log10 + clip (robust)
# -----------------------
for c in keep_cols:
    # log10(x) (x  >0)
    gdf_points[c] = np.log10(gdf_points[c].astype(float))

    # clip outlier (0.5% - 99.5%)  trnh survey extreme dominate
    lo = gdf_points[c].quantile(0.005)
    hi = gdf_points[c].quantile(0.995)
    gdf_points[c] = gdf_points[c].clip(lo, hi)

# -----------------------
# 6) Grid 500m + aggregate features
# -----------------------
from shapely.geometry import box

CELL = 500

xmin, ymin, xmax, ymax = gdf_points.total_bounds
xs = np.arange(xmin, xmax, CELL)
ys = np.arange(ymin, ymax, CELL)

cells = []
cell_ids = []
k = 0
for x in xs:
    for y in ys:
        cells.append(box(x, y, x+CELL, y+CELL))
        cell_ids.append(k)
        k += 1

grid = gpd.GeoDataFrame({"cell_id": cell_ids}, geometry=cells, crs=gdf_points.crs)

# spatial join
joined = gpd.sjoin(
    gdf_points[["geometry", "SAMPLEID", "COMPANY"] + keep_cols],
    grid,
    how="left",
    predicate="within"
)

# bias controls
bias = joined.groupby("cell_id").agg(
    n_samples=("SAMPLEID", "count"),
    n_company=("COMPANY", pd.Series.nunique),
)

# aggregate elements
gb = joined.groupby("cell_id")[keep_cols]
feat = gb.agg(["median", lambda s: s.quantile(0.9), "max", "count"])
feat.columns = [f"{c}__{stat if isinstance(stat,str) else 'p90'}" for c, stat in feat.columns]

grid_ml = grid.merge(feat, left_on="cell_id", right_index=True, how="left").merge(
    bias, left_on="cell_id", right_index=True, how="left"
)

print("Grid ML shape:", grid_ml.shape)

# -----------------------
# 7) Save
# -----------------------
```

```python
import numpy as np
import pandas as pd

trace_cols = ["Cu","Zn","Pb","Ag","Ni","Co","Cr","V","Mo","As","Sb","Bi","Au"]

NULL_CODE = -9999  # theo dictionary

def clean_trace(df, col):
    s = pd.to_numeric(df[col], errors="coerce")

    # 1) NULL -> NaN
    s = s.mask(s == NULL_CODE, np.nan)

    # 2) BDL flag: m nhng khng phi NULL
    is_bdl = (s < 0) & (~s.isna())

    # 3) Impute BDL: -x ngha l "< x"  => thay bng 0.5*x
    s_clean = s.copy()
    s_clean.loc[is_bdl] = 0.5 * np.abs(s.loc[is_bdl])

    # 4) Nu cn m l t do li parse -> set NaN
    s_clean = s_clean.mask(s_clean < 0, np.nan)

    # 5) Log feature (trace elements thng rt skewed)
    s_log = np.log1p(s_clean)

    return s_clean, is_bdl.astype("int8"), s_log

for c in trace_cols:
    if c in gdf_ozchem.columns:
        gdf_ozchem[f"{c}_clean"], gdf_ozchem[f"{c}_is_bdl"], gdf_ozchem[f"{c}_log"] = clean_trace(gdf_ozchem, c)
```

```python
import numpy as np
import pandas as pd
import geopandas as gpd
from shapely.geometry import box

# =========================
# 0) SETTINGS
# =========================
CELL_SIZE = 500  # meters
CRS = "EPSG:28352"

# =========================
# 1) SANITY CHECK CRS
# =========================
if gdf_ozchem.crs is None:
    raise ValueError("gdf_ozchem.crs is None. Please set CRS first (EPSG:28352).")

if str(gdf_ozchem.crs).upper() != CRS.upper():
    gdf_ozchem = gdf_ozchem.to_crs(CRS)

# (Optional) keep only points with valid geometry
gdf_ozchem = gdf_ozchem[gdf_ozchem.geometry.notna()].copy()

# =========================
# 2) BUILD GRID 500m
#    (grid covers the extent of OZCHEM points; you can later replace it with your master NT grid)
# =========================
xmin, ymin, xmax, ymax = gdf_ozchem.total_bounds

# snap bounds to grid and add a 1-cell padding
xmin = np.floor(xmin / CELL_SIZE) * CELL_SIZE - CELL_SIZE
ymin = np.floor(ymin / CELL_SIZE) * CELL_SIZE - CELL_SIZE
xmax = np.ceil(xmax / CELL_SIZE) * CELL_SIZE + CELL_SIZE
ymax = np.ceil(ymax / CELL_SIZE) * CELL_SIZE + CELL_SIZE

xs = np.arange(xmin, xmax, CELL_SIZE)
ys = np.arange(ymin, ymax, CELL_SIZE)

polys = []
cell_ids = []
rows = []
cols = []

cell_id = 0
for r, y in enumerate(ys):
    for c, x in enumerate(xs):
        polys.append(box(x, y, x + CELL_SIZE, y + CELL_SIZE))
        cell_ids.append(cell_id)
        rows.append(r)
        cols.append(c)
        cell_id += 1

grid = gpd.GeoDataFrame(
    {"cell_id": cell_ids, "row": rows, "col": cols},
    geometry=polys,
    crs=CRS
)

# Add centroid coordinates (useful for ML / exporting)
grid["cell_x"] = grid.geometry.centroid.x
grid["cell_y"] = grid.geometry.centroid.y

# =========================
# 3) SPATIAL JOIN: points -> grid cell_id
# =========================
# Keep only columns we need to aggregate (for speed)
oxide_cols = ["SiO2", "TiO2", "MgO", "CaO", "Na2O", "K2O", "P2O5", "Fe2O3tot_filled"]

trace_core = ["Cu_log", "Mo_log", "As_log", "Zn_log", "Pb_log"]
trace_support = ["Ag_log", "Sb_log", "Bi_log", "Au_log", "Ni_log", "Co_log", "Cr_log", "V_log"]

bdl_flags = ["Cu_is_bdl", "Mo_is_bdl", "As_is_bdl", "Ag_is_bdl", "Bi_is_bdl", "Au_is_bdl"]

# only keep columns that exist
use_cols = []
for lst in [oxide_cols, trace_core, trace_support, bdl_flags]:
    use_cols.extend([c for c in lst if c in gdf_ozchem.columns])

# Include an ID column for counting samples (fallback to index if no uid)
id_col = "uid" if "uid" in gdf_ozchem.columns else None
if id_col:
    pts = gdf_ozchem[[id_col] + use_cols + ["geometry"]].copy()
else:
    pts = gdf_ozchem[use_cols + ["geometry"]].copy()
    pts["_tmp_id"] = np.arange(len(pts))
    id_col = "_tmp_id"

# spatial join
pts_join = gpd.sjoin(
    pts,
    grid[["cell_id", "geometry"]],
    how="left",
    predicate="within"
)

# drop points that fall outside the grid (should be none)
pts_join = pts_join.dropna(subset=["cell_id"]).copy()
pts_join["cell_id"] = pts_join["cell_id"].astype(int)

# =========================
# 4) AGGREGATION FUNCTIONS
# =========================
def q90(x):
    return x.quantile(0.90)

def q95(x):
    return x.quantile(0.95)

# Build aggregation dict
agg_dict = {}

# ---- Core trace metals: mean + p90 + max
for c in [col for col in trace_core if col in pts_join.columns]:
    agg_dict[c] = ["mean", q90, "max"]

# ---- Support trace metals:
# Default: mean + p90
for c in [col for col in trace_support if col in pts_join.columns]:
    if c == "Bi_log":
        agg_dict[c] = [q90, "max"]          # Bi: focus on high tail
    elif c == "Au_log":
        agg_dict[c] = [q90]                 # Au: p90 only (often tiny)
    else:
        agg_dict[c] = ["mean", q90]

# ---- BDL flags: rate (mean) + count (sum)
for c in [col for col in bdl_flags if col in pts_join.columns]:
    agg_dict[c] = ["mean", "sum"]

# ---- Oxides: mean (+ optional std)
for c in [col for col in oxide_cols if col in pts_join.columns]:
    agg_dict[c] = ["mean", "std"]

# Overall sample count per cell
# We'll aggregate separately for clarity
grouped = pts_join.groupby("cell_id")

agg_df = grouped.agg(agg_dict)
agg_df["sample_count"] = grouped[id_col].size()

# Flatten multi-index columns
new_cols = []
for col in agg_df.columns:
    if isinstance(col, tuple):
        base, stat = col
        if callable(stat):
            stat_name = stat.__name__
        else:
            stat_name = str(stat)

        # rename quantile function to q90, std, etc.
        if stat_name == "<lambda>":
            stat_name = "func"

        new_cols.append(f"{base}_{stat_name}")
    else:
        new_cols.append(col)

agg_df.columns = new_cols
agg_df = agg_df.reset_index()

# Rename BDL outputs to clearer names
# e.g., Mo_is_bdl_mean -> Mo_is_bdl_rate, Mo_is_bdl_sum -> Mo_is_bdl_count
for c in bdl_flags:
    if f"{c}_mean" in agg_df.columns:
        agg_df = agg_df.rename(columns={f"{c}_mean": f"{c}_rate"})
    if f"{c}_sum" in agg_df.columns:
        agg_df = agg_df.rename(columns={f"{c}_sum": f"{c}_count"})

# Rename q90 columns to p90 (more readable)
agg_df.columns = [col.replace("_q90", "_p90") for col in agg_df.columns]

# =========================
# 5) MERGE BACK TO GRID (keep all cells, even empty ones)
# =========================
grid_feat = grid.merge(agg_df, on="cell_id", how="left")

# For empty cells, sample_count should be 0 instead of NaN
if "sample_count" in grid_feat.columns:
    grid_feat["sample_count"] = grid_feat["sample_count"].fillna(0).astype(int)

# =========================
# 6) EXPORT
# =========================
# 6.1 Grid with geometry + features (great for QGIS)
# grid_feat.to_file("ozchem_grid_500m.gpkg", layer="ozchem_grid_500m", driver="GPKG")

# 6.2 ML table without geometry (faster for ML)
ml_df = pd.DataFrame(grid_feat.drop(columns=["geometry"]))
# ml_df.to_parquet("ozchem_features_500m.parquet", index=False)

print("Done!")
print("Saved: ozchem_grid_500m.gpkg")
print("Saved: ozchem_features_500m.parquet")
```
